# Modèle de Black–Scholes
<br>

*Ce notebook a pour objectif de présenter le modèle de Black–Scholes et d’étudier l’évolution de ses Greeks lorsque certains paramètres varient.
Nous commencerons par exposer l’approche théorique du modèle, puis nous implémenterons la formule de Black–Scholes sous la forme d’une classe capable de calculer les différents greeks. Enfin, nous analyserons l’évolution de ces sensibilités en fonction des paramètres clés du modèle.*

### Solution fermée pour le prix du sous-jacent

Dans le modèle de Black–Scholes, si le prix initial du sous-jacent est \$ S_0 \$, alors sa valeur à la date \$ t \$ est donnée par :

$$
S_t = S_0 \exp\left[ \left( \mu - \frac{1}{2}\sigma^2 \right) t + \sigma W_t \right]
$$

où \$ \mu \$ est le taux de croissance attendu de l’actif, \$ \sigma \$ la volatilité constante et \$ W_t \$ un mouvement brownien standard.

Sous la mesure risque-neutre \$ \mathbb{Q} \$ :

$$
S_t = S_0 \exp\left[ \left( r - \frac{1}{2}\sigma^2 \right) t + \sigma W_t^{\mathbb{Q}} \right]
$$

### Dynamique différentielle

Le processus suivi par \$ S_t \$ s’écrit :

$$
dS_t = \mu S_t\, dt + \sigma S_t\, dW_t
$$

et sous \$ \mathbb{Q} \$ :

$$
dS_t = r S_t\, dt + \sigma S_t\, dW_t^{\mathbb{Q}}
$$

### Prix d'une option européenne

En posant \$ \tau = T - t \$ le temps restant jusqu'à l’échéance :

Call européen :

$$
C(S, K, r, \sigma, \tau) = S\, N(d_1) - K e^{-r\tau} N(d_2)
$$

Put européen :

$$
P(S, K, r, \sigma, \tau) = K e^{-r\tau} N(-d_2) - S\, N(-d_1)
$$

avec :

$$
d_1 = \frac{\ln\left( \frac{S}{K} \right) + \left( r + \frac{1}{2}\sigma^2 \right) \tau}{\sigma \sqrt{\tau}}, 
\quad
d_2 = d_1 - \sigma \sqrt{\tau}
$$

### Delta

Call :

$$
\Delta_C = \frac{\partial C}{\partial S} = N(d_1)
$$

Put :

$$
\Delta_P = \frac{\partial P}{\partial S} = N(d_1) - 1 = -N(-d_1)
$$

### Gamma

$$
\Gamma = \frac{\partial^2 C}{\partial S^2} = \frac{\partial^2 P}{\partial S^2} = \frac{N'(d_1)}{S \sigma \sqrt{\tau}}
$$

avec :

$$
N'(d_1) = \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} d_1^2}
$$

### Vega

$$
\mathcal{V} = \frac{\partial C}{\partial \sigma} = \frac{\partial P}{\partial \sigma} = S\, N'(d_1) \sqrt{\tau}
$$

### Theta

Call :

$$
\Theta_C = \frac{\partial C}{\partial \tau} = -\frac{S N'(d_1)\sigma}{2\sqrt{\tau}} - r K e^{-r\tau} N(d_2)
$$

Put :

$$
\Theta_P = \frac{\partial P}{\partial \tau} = -\frac{S N'(d_1)\sigma}{2\sqrt{\tau}} + r K e^{-r\tau} N(-d_2)
$$

### Rho

Call :

$$
\rho_C = \frac{\partial C}{\partial r} = K \tau e^{-r\tau} N(d_2)
$$

Put :

$$
\rho_P = \frac{\partial P}{\partial r} = -K \tau e^{-r\tau} N(-d_2)
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

class BS:
    def __init__(self, S, K, r, T, sigma, optype="C"):
        if optype not in ("C", "P"):
            raise ValueError("call ou put option")
        if S <= 0 or K <= 0 or T <= 0 or sigma <= 0:
            raise ValueError("S, K, T, sigma doivent être strictement positifs")
        self.S = float(S)
        self.K = float(K)
        self.r = float(r)
        self.T = float(T)
        self.sigma = float(sigma)
        self.optype = optype

    @property
    def d1(self):
        return (np.log(self.S / self.K)+ (self.r + 0.5 * self.sigma**2) * self.T) / (self.sigma * np.sqrt(self.T))

    @property
    def d2(self):
        return self.d1 - self.sigma * np.sqrt(self.T)

    def price(self):
        if self.optype == "C":
            return self.S * norm.cdf(self.d1) - self.K * np.exp(-self.r * self.T) * norm.cdf(self.d2)
        else:
            return self.K * np.exp(-self.r * self.T) * norm.cdf(-self.d2) - self.S * norm.cdf(-self.d1)

    def delta(self):
        if self.optype == "C":
            return norm.cdf(self.d1)
        else:
            return -norm.cdf(-self.d1)

    def gamma(self):
        return norm.pdf(self.d1) / (self.S * self.sigma * np.sqrt(self.T))

    def vega(self):
        return self.S * norm.pdf(self.d1) * np.sqrt(self.T)

    def theta(self):
        if self.optype == "C":
            return -self.S * norm.pdf(self.d1) * self.sigma / (2 * np.sqrt(self.T)) - self.r * self.K * np.exp(-self.r * self.T) * norm.cdf(self.d2)
        else:
            return -self.S * norm.pdf(self.d1) * self.sigma / (2 * np.sqrt(self.T)) + self.r * self.K * np.exp(-self.r * self.T) * norm.cdf(-self.d2)

    def rho(self):
        if self.optype == "C":
            return self.K * self.T * np.exp(-self.r * self.T) * norm.cdf(self.d2)
        else:
            return -self.K * self.T * np.exp(-self.r * self.T) * norm.cdf(-self.d2)
```

# Greeks Study

```python
K = 1000.0
r = 0.01
sigma = 0.2
optype = "C"

tau_values = [1/365, 7/365, 30/365, 60/365, 0.5]
labels = ["1 jour", "1 semaine", "1 mois", "2 mois", "6 mois"]
S_values = np.linspace(900, 1100, 200)

plt.figure(figsize=(8,5))

for idx, tau in enumerate(tau_values):
    deltas = [BS(S, K, r, tau, sigma, optype).delta() for S in S_values]
    plt.plot(S_values, deltas, color=str(0.9 - 0.8 * idx / len(tau_values)), linewidth=2, label=labels[idx])

plt.title("Évolution du Delta en fonction de S pour différentes maturités")
plt.xlabel("Sous-jacent S")
plt.ylabel("Delta")
plt.legend(title="Maturité")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))

for idx, tau in enumerate(tau_values):
    gammas = [BS(S, K, r, tau, sigma, optype).gamma() for S in S_values]
    plt.plot(S_values, gammas, color=str(0.9 - 0.8 * idx / len(tau_values)), linewidth=2, label=labels[idx])

plt.title("Évolution du Gamma en fonction de S pour différentes maturités")
plt.xlabel("Sous-jacent S")
plt.ylabel("Gamma")
plt.legend(title="Maturité")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))

for idx, tau in enumerate(tau_values):
    thetas = [BS(S, K, r, tau, sigma, optype).theta() for S in S_values]
    plt.plot(S_values, thetas, color=str(0.9 - 0.8 * idx / len(tau_values)), linewidth=2, label=labels[idx])

plt.title("Évolution du Theta en fonction de S pour différentes maturités")
plt.xlabel("Sous-jacent S")
plt.ylabel("Theta")
plt.legend(title="Maturité")
plt.grid(True)
plt.show()
```

```python
S_values = np.linspace(600, 1600, 200)
plt.figure(figsize=(8,5))

for idx, tau in enumerate(tau_values):
    vegas = [BS(S, K, r, tau, sigma, optype).vega() for S in S_values]
    plt.plot(S_values, vegas, color=str(0.9 - 0.8 * idx / len(tau_values)), linewidth=2, label=labels[idx])

plt.title("Évolution du Vega en fonction de S pour différentes maturités")
plt.xlabel("Sous-jacent S")
plt.ylabel("Vega")
plt.legend(title="Maturité")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))

for idx, tau in enumerate(tau_values):
    rhos = [BS(S, K, r, tau, sigma, optype).rho() for S in S_values]
    plt.plot(S_values, rhos, color=str(0.9 - 0.8 * idx / len(tau_values)), linewidth=2, label=labels[idx])

plt.title("Évolution du Rho en fonction de S pour différentes maturités")
plt.xlabel("Sous-jacent S")
plt.ylabel("Rho")
plt.legend(title="Maturité")
plt.grid(True)
plt.show()
```

# Greeks surface

```python
from mpl_toolkits.mplot3d import Axes3D

S0 = 1000.0   
r = 0.01       
sigma = 0.2    
optype = "C" 

K_values=np.linspace(800,1200,60)
tau_values=np.linspace(1/365,0.5,60)

Delta=np.zeros((len(tau_values),len(K_values)))

for i , tau in enumerate(tau_values):
    for j, K in enumerate(K_values):
        option=BS(S0,K,r,tau,sigma,optype)
        Delta[i,j]=option.delta()

K_mesh,tau_mesh=np.meshgrid(K_values,tau_values)
fid=plt.figure(figsize=(9,6))
ax=fid.add_subplot(111,projection='3d')
ax.plot_surface(K_mesh,tau_mesh,Delta, cmap='viridis')

ax.set_title("Call option delta surface")
ax.set_xlabel("Strike (K)")
ax.set_ylabel(r"$\tau$ (time to expiration)")
ax.set_zlabel("Delta")

plt.show()
```

```python
Gamma = np.zeros((len(tau_values), len(K_values)))

for i, tau in enumerate(tau_values):
    for j, K in enumerate(K_values):
        option = BS(S0, K, r, tau, sigma, optype)
        Gamma[i, j] = option.gamma()

K_mesh, tau_mesh = np.meshgrid(K_values, tau_values)
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(K_mesh, tau_mesh, Gamma, cmap='viridis')

ax.set_title("Call option gamma surface")
ax.set_xlabel("Strike (K)")
ax.set_ylabel(r"$\tau$ (time to expiration)")
ax.set_zlabel("Gamma")

plt.show()
```

```python
Theta=np.zeros((len(tau_values),len(K_values)))

for i , tau in enumerate(tau_values):
    for j, K in enumerate(K_values):
        option=BS(S0,K,r,tau,sigma,optype)
        Theta[i,j]=option.theta()

K_mesh,tau_mesh=np.meshgrid(K_values,tau_values)

fig=plt.figure(figsize=(9,6))
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(K_mesh,tau_mesh,Theta, cmap='viridis')

ax.set_title("Call option Theta surface")
ax.set_xlabel("Strike(K)")
ax.set_ylabel(r"$\tau$ (time to expiration)")
ax.set_zlabel("Theta")

plt.show()
```

```python
Vega=np.zeros((len(tau_values),len(K_values)))

for i , tau in enumerate(tau_values):
    for j, K in enumerate(K_values):
        option=BS(S0,K,r,tau,sigma,optype)
        Vega[i,j]=option.vega()

K_mesh,tau_mesh=np.meshgrid(K_values,tau_values)

fig=plt.figure(figsize=(9,6))
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(K_mesh,tau_mesh,Vega, cmap='viridis')

ax.set_title("Call option Vega surface")
ax.set_xlabel("Strike(K)")
ax.set_ylabel(r"$\tau$ (time to expiration)")
ax.set_zlabel("Vega")

plt.show()
```

```python
Rho=np.zeros((len(tau_values),len(K_values)))

for i , tau in enumerate(tau_values):
    for j, K in enumerate(K_values):
        option=BS(S0,K,r,tau,sigma,optype)
        Rho[i,j]=option.rho()

K_mesh,tau_mesh=np.meshgrid(K_values,tau_values)

fig=plt.figure(figsize=(9,6))
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(K_mesh,tau_mesh,Rho, cmap='viridis')

ax.set_title("Call option Rho surface")
ax.set_xlabel("Strike(K)")
ax.set_ylabel(r"$\tau$ (time to expiration)")
ax.set_zlabel("Rho")

plt.show()
```
